```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def solve_le(n):
    


    def f(t, y):
        a= n
        dydt = [y[1] , (-2/t)*y[1]-y[0]**(a)]  
        return dydt

    dt = 0.000001
    tspan = np.linspace(0+dt,20,1000)  
    yinit = [1,0]
    
    sol = solve_ivp(lambda t, y: f(t, y),[tspan[0], tspan[-1]], yinit,method='RK45', t_eval=tspan)

    return sol
```

```python
n= [1,1.5,2,3,4,5]
fig, axis = plt.subplots(figsize = (9,5))
for i in n:
    sol = solve_le(i)
    plt.plot(sol.t,sol.y[0])
plt.ylim((0,1.01))
plt.xlim((0,20))
plt.gca().legend(('n=1','n=1.5','n=2','n=3','n=4','n=5'), loc='upper right')
plt.title("solutions of Lane-Emden Equation for different n values")
plt.xlabel("Scaled Radius")
plt.ylabel("Dimensionless Density")
```
